```python
# Building a modularized function for year and URL
year = '2022'
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

url = str.format(year)
url
```

```python
# for multiple years, we can use a loop
years = [2022,2023,2024]
url_link = 'https://www.basketball-reference.com/leagues/NBA_{}_per_game.html'

for year in years:
  url = str.format(url_link, year)
  print(url)
```

```python
# Read HTML page using Pandas
import pandas as pd

df = pd.read_html(url, header=0)
df
```
